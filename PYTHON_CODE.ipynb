```python
filename = input("Enter the valu of filename: ")
f = open(filename, 'r')
text = f.read()
text = text.replace(" ","")
alphabet_count = {}
for character in text:
    if character.isalpha():
        if character in alphabet_count:
          alphabet_count[character] += 1
        else:
            alphabet_count[character] = 1
print("Alphabet Count:")
for character, count in alphabet_count.items():
    print(character, ":", count)
f.close()
```
